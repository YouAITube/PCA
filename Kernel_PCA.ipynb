```python
from matplotlib.colors import ListedColormap, to_rgba
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles

def plot_data(ax, ds, name='', marker='o', colors=("red", "purple"), alpha=0.7, s=150, facecolor=True):
    X, y = ds

    # Convert color names to RGBA
    colors_rgba = [to_rgba(color) for color in colors]
    cm_bright = ListedColormap(colors_rgba)

    ax.set_title(name, fontsize=30)
    if facecolor:
        ax.scatter(X[:, 0], X[:, 1], marker=marker, c=y, cmap=cm_bright, s=s, alpha=alpha)
    else:
        ax.scatter(X[:, 0], X[:, 1], marker=marker, c=y, cmap=cm_bright, s=s, facecolor='None', alpha=alpha)

    x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    ax.grid(True)
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)
    ax.set_xticks(())
    ax.set_yticks(())

```

```python
# Generate data and plot
N = 150
noises = 0.05
random_state = 42

X_circle, y_circle = make_circles(n_samples=N, noise=noises, factor=0.5, random_state=random_state)
fig, ax = plt.subplots(1, 1)
plot_data(ax, (X_circle, y_circle))
plt.show()
```

```python
from sklearn.decomposition import  KernelPCA
```



* the number of `n_components` (which we want to keep in the end)
* the type of kernel `kernel` and the kernel hyperparameters
  * `' poly'`: polynomial kernel.
      * `' degree '` : degree of a polynomial

*   List item
*   List item



 * `'rbf'`: Radial basis function.
     * `' gamma '` : radial function parameter




```python
kernel = 'rbf'  #@param [ 'rbf' , 'sigmoid', 'poly']{type:"string"}
coef0=2 #@param {type:"slider", min:0, max:5, step:0.5}
degree=3 #@param {type:"slider", min:1, max:3, step:1}
gamma=10 #@param {type:"slider", min:0.00, max:20, step:0.5}
#@markdown ### number of components
n_components=0 #@param {type:"slider", min:0, max:100, step:1}


if n_components == 0:
   n_components=None

kernel_pca = KernelPCA(
    n_components=n_components, kernel=kernel,
    degree = degree, coef0 = coef0,
    gamma=gamma, fit_inverse_transform=True, alpha=0.1
)
```

```python
kernel_pca.fit(X_circle)
```

```python
kpca_result = kernel_pca.transform(X_circle)
```

```python
x_component=5 #@param {type:"slider", min:1, max:9, step:1}
y_component=6 #@param {type:"slider", min:1, max:9, step:1}

fig = plt.figure(figsize=(9,9))

plt.scatter(kpca_result[:,x_component-1],
            kpca_result[:,y_component-1],
            alpha = 0.6, s = 100,
            c = y_circle,
            cmap = 'bwr');
plt.xticks([],[])
plt.yticks([],[])
```
